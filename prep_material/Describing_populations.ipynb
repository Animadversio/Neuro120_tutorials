## Describing the spread or variability of the population
### What is the range of the Martian heights?
### What is the height of the shortest and tallest Martians?

```python
short_m = np.amin(mar_h)
tall_m = np.amax(mar_h)
range_m = tall_m - short_m
print(short_m)
print(tall_m)
print(range_m)
```

## Percentiles
### Another way to describe the spread of the population is to split it into percentages
### How tall is the first 25% of Martians? 50%? 75%?
### These are Quartiles which are commonly used
### Quantiles are the same thing without converting into percent

```python
Q1_mar = np.percentile(mar_h, 25)
Q2_mar = np.percentile(mar_h, 50)
Q3_mar = np.percentile(mar_h, 75)
print(Q1_mar)
print(Q2_mar)
print(Q3_mar)
```

```python
plt.title("25th, 50th, and 75th percentiles of Martian height")
plt.xlabel("Height (cm)")
plt.ylabel("Count")
_=plt.hist(mar_h,bins=100) #
plt.axvline(Q1_mar, color='r', linestyle='dotted', linewidth=2)
plt.axvline(Q2_mar, color='r', linestyle='dotted', linewidth=2)
plt.axvline(Q3_mar, color='r', linestyle='dotted', linewidth=2)
```

## Interquartile range (IQR) is another measure of population spread or dispersion
### Spread of the middle 50% of the population
### Difference between the 75th and 25th percentiles

```python
mar_h_iqr= np.percentile(mar_h, 75)-np.percentile(mar_h, 25)
print (mar_h_iqr)
```

## How variable are these descriptions of the population?
### Some measure are more stable than others
### Range (largest - smallest) relies on two values so will be more variable in comparison to mean, quartiles, and IQR.

```python
test = np.random.normal(10,2,size=1000)
test = np.round(test,2)
test_range= np.amax(test)-np.amin(test)
print(test_range)
test_mean = np.mean(test)
print(test_mean)
test_iqr= np.percentile(test, 75)-np.percentile(test, 25)
print (test_iqr)
```

## Defining population variability
### What is the dispersion of each Martian's height around the mean?
### Calculate the difference for each Martian from the mean.

```python
diff_m = (mar_h-np.mean(mar_h))
plt.title("Histogram of difference between each Martian's height and population mean")
_=plt.hist(diff_m, bins=100)
```

## Variance (2] ) = sum of squared differences / number
### $$ \sigma^2 = \mathrm{var}(x_1, \ldots, x_2) = \frac{1}{n} \sum_i (x_i - \mu)^2 $$
### Squaring the differences eliminates the negative numbers

```python
mar_var = np.var(mar_h)
print(mar_var)
```

## Standard deviation () = square root of the variance
### $$ \sigma = \sqrt{\mathrm{var}(x_1, \ldots, x_n)} $$
### Often abbreviated SD or s.d.
### Variance has squared units that are not intuitive
### SD have same units as original measures


```python
mar_std1 = np.sqrt(np.var(mar_h))
mar_std2 = np.std(mar_h)
print (mar_std1)
print (mar_std2)
```

```python

plt.title("1 and 2 Standard Deviations ")
plt.xlabel("Height (cm)")
plt.ylabel("Count")
_=plt.hist(mar_h,bins=100) #
plt.axvline(mean_m, color='k', linestyle='dashed', linewidth=2)
plt.axvline((mean_m + mar_std1), color='r', linestyle='dotted', linewidth=2)
plt.axvline((mean_m + 2*mar_std1), color='r', linestyle='dotted', linewidth=2)
plt.axvline((mean_m - mar_std1), color='r', linestyle='dotted', linewidth=2)
plt.axvline((mean_m - 2*mar_std1), color='r', linestyle='dotted', linewidth=2)
```

```python
## SD is related to percentiles
#1 SD = 16 and 84, or 68% of the population
#2 SD = 2.5 and 97.5, or 95%.
plt.title("Percentiles (2.5, 16, 84, and 97.5) of Martian Height")
plt.xlabel("Height (cm)")
plt.ylabel("Count")
_=plt.hist(mar_h, bins=100) #
plt.axvline(mean_m, color='k', linestyle='dashed', linewidth=2)
plt.axvline(np.percentile(mar_h, 84), color='r', linestyle='dotted', linewidth=2)
plt.axvline(np.percentile(mar_h, 16), color='r', linestyle='dotted', linewidth=2)
plt.axvline(np.percentile(mar_h, 2.5), color='r', linestyle='dotted', linewidth=2)
plt.axvline(np.percentile(mar_h, 97.5), color='r', linestyle='dotted', linewidth=2)
```

## Sampling a population
### Budget cuts!!! You couldn't send the entire team to Venus.
### You can only measure some of the Venusians, not the entire population.

```python
#Mean for subset of population
ven_h_sub100=np.random.choice(ven_h, size = 100)
mean_ven_h_sub100=np.mean(ven_h_sub100)
print(mean_ven_h_sub100)
print(mean_v)
```

## Variance and SD for a sample is different than the entire population
### Based on n - 1, rather than n
### Added degree of freedom for variability in sampling

```python
ven_std_sample=np.std(ven_h_sub100, ddof=1)
ven_std_pop=np.std(ven_h_sub100)
print(ven_std_sample)
print(ven_std_pop)
```

### Random sampling of subsets of population
### What will the distribution of means of these samples look like?
### Central Limit Theorem
### Mean of means is gaussian, even from non-gaussian populations

```python
# Plot the histogram of multiple random samples of venusian distribution
ven_h_sub2000=np.random.choice(ven_h, size = 2000) #random sample of 2000
ven_h_array=(ven_h_sub2000.reshape(40,50)) #split the 2000 into 40 X 50 to mimic sampling 50 Venusians forty times.
ven_sample_means=np.mean(ven_h_array,axis=1) #means of those Venusian samples
_=plt.hist(ven_sample_means)
ven_sample_means=np.mean(ven_sample_means)
print(ven_sample_means)
print(mean_v)
```
